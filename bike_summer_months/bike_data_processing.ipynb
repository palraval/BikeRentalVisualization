```python
# reading the summer month data 

import pandas as pd

june_data = pd.read_csv('~/Downloads/bike_summer_months/JC-202306-citibike-tripdata.csv')

july_data = pd.read_csv('~/Downloads/bike_summer_months/JC-202307-citibike-tripdata.csv')

august_data = pd.read_csv('~/Downloads/bike_summer_months/JC-202308-citibike-tripdata.csv')

september_data = pd.read_csv('~/Downloads/bike_summer_months/JC-202309-citibike-tripdata.csv')

```

```python
d = pd.read_csv('bike_original_data/bike_summer_months/JC-202306-citibike-tripdata.csv')

d.head()
```

```python
# combining separate dataframes into one total dataframe

summer_data = pd.concat([june_data, july_data, august_data, september_data], axis = 0)

summer_data.to_csv('~/Downloads/bike_summer_months/summer_data.csv')
```

```python
summer_data.shape
```

```python
# creating dataframe for counts of each start station in june

june_station_count = june_data.groupby(['start_station_name'])['start_station_name'].count()

june_station_count_df = pd.DataFrame(june_station_count)


june_station_count_df.columns = ['start station popularity']

june_station_count_df = june_station_count_df.reset_index()

# adding column on june date

june_station_count_df['data info'] = "2023-06-01"

june_station_count_df.head()
```

```python
# creating dataframe for counts of each start station in july

july_station_count = july_data.groupby(['start_station_name'])['start_station_name'].count()

july_station_count_df = pd.DataFrame(july_station_count)


july_station_count_df.columns = ['start station popularity']

july_station_count_df = july_station_count_df.reset_index()

# adding column on july date

july_station_count_df['data info'] = "2023-07-01"

july_station_count_df.head()
```

```python
# creating dataframe for counts of each start station in august

august_station_count = august_data.groupby(['start_station_name'])['start_station_name'].count()

august_station_count_df = pd.DataFrame(august_station_count)


august_station_count_df.columns = ['start station popularity']

august_station_count_df = august_station_count_df.reset_index()

# adding column on august date

august_station_count_df['data info'] = "2023-08-01"

august_station_count_df.head()
```

```python
# creating dataframe for counts of each start station in september

september_station_count = september_data.groupby(['start_station_name'])['start_station_name'].count()

september_station_count_df = pd.DataFrame(september_station_count)


september_station_count_df.columns = ['start station popularity']

september_station_count_df = september_station_count_df.reset_index()

# adding column on september date

september_station_count_df['data info'] = "2023-09-01"

september_station_count_df.head()
```

```python
# combining all the count dataframes into a single dataframe and converts to csv

summer_station_count_data = pd.concat([june_station_count_df, july_station_count_df,august_station_count_df,
                                       september_station_count_df], axis = 0)

summer_station_count_data.to_csv('~/Downloads/bike_summer_months/summer_station_count_data.csv')
```
